```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import re
import seaborn as sns
%matplotlib inline
```

```python
video = pd.read_csv('E:\Data Science Learning\Data Analytics\Youtube/USvideos.csv',error_bad_lines=False)
```

```python
video.head()
```

## Task 1 : Analysing Tags column to check what are trending tags on youtube?

```python
tags_total = ' '.join(video['tags'])

tags_total
```

```python
re.sub('[^a-zA-Z]+',' ',tags_total)
```

```python
wordcloud = WordCloud(width=1500,height=500,stopwords=set(STOPWORDS)).generate(tags_total)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis('off')
```

## Task 2 : Performing Analysis on likes, views & Dislikes to find how they are co-related?

```python
sns.regplot(data=video,x='views',y='likes')
plt.title('Effect of views on likes')
```

```python
sns.regplot(data=video,x='views',y='dislikes')
plt.title('Effect of views on dislikes')
```

```python
df_corr = video[['views','likes','dislikes']]
```

```python
df_corr.corr()
```

```python
sns.heatmap(df_corr.corr(),annot=True)
```
